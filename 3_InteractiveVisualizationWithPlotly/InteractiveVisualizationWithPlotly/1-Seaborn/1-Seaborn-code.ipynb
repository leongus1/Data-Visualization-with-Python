```python
#=================================================-
#### Slide 8: Loading the dataset and libraries  ####

# import the libraries
import seaborn as sns
from matplotlib import pyplot as plt

# Load the dataset
penguins = sns.load_dataset("penguins")

# Top 5 entries of dataset
penguins.head()


```

```python
#=================================================-
#### Slide 9: Data preprocessing  ####

# Check for null values
penguins.isna().sum()
# Drop NA's and reassign clean data
penguins = penguins.dropna()


```

```python
#=================================================-
#### Slide 23: Multivariate plots: Heatmaps  ####
import numpy as np
plt.figure(figsize=(6,4))            #<- set the figure size+
corr = penguins.select_dtypes("number").corr()
sns.heatmap(
    corr.iloc[1:, :-1],
    center=0,
    annot=True,
    square=True,
    mask=np.triu(corr)[1:,:-1],
    annot_kws={"fontsize":14},
    vmin = -1, 
    vmax = 1, 
)


```

## Static plot type guidelines
- timestamped data points? 1 valiable is time? => line plot (if too wide, maybe do a facet grid or years as separate lines with confidence bands )
- proportions out of a whole? then group bar plot sorted by value (dont use stacked bar, treemap, pie chart, or donut chart)(instead of small multiples, try a sankey / alluvial diagram)
- univariate distribution? density plot (kde), rugplot, ecdf, histogram, violin, boxplot, beeswarm(jitter plot)
- single value per group? horizontal bar plot sorted by value
- relationship between 2 variables? scatter plot(maybe add a contour on top if want to emphasize a general pattern rathen than individual points) 
    - dont do 3D for static plots, instead use marker shape, color, size for additional dimension/variables
    - use a jointplot to combine univariate plots with your scatter plot
- geospacial data? use choropleth or bubble chart on top of a map
- crosstab? heatmap (perhaps with dendograms in the margins)
