## What you want to visualize determines the plot type
```mermaid
flowchart TB

A{{Univariate?}} --> B[Actual values?] --> D(Histogram/Density/Rug)
    B --> F(Violin/Box/Swarm)
A --> G[Proportions?] --> H(Grouped Bar)
A --> R[Summary values?] --> S(Bar)
    I{{Multivariate?}} --> J[Relationship?] --> K(Scatter/Contour)
    I --> N[Over time?] --> O(Line)
    I --> P[Across locations?] --> Q(Choropleth)
    I --> T[Tabular?] --> U(Heatmap)

## Static plot type guidelines
- timestamped data points? 1 variable is time? => line plot with time on x-axis (if it is too wide, have years as separate lines; if too many lines, do a facet grid)
- proportions out of a whole? grouped bar plot sorted by value (not stacked bar, treemap, pie chart, or donut chart) (instead of small multiples, try a sankey / alluvial / ribbon diagram)
- univariate distribution? density plot (kde), rugplot, ecdf, histogram, violin, boxplot, beeswarm (jitter plot), strip plot, parallel coordinates plot
- single value per group? horizontal bar plot sorted by value
- relationship between 2 variables? scatter plot (maybe add a contour on top if want to emphasize a general pattern rather than individual points)
    -don't do 3d static plots, instead use marker shape, color, and size for additional dimensions/variables
    -use a jointplot to add univariate plots to the margins of your scatter plot
- geospatial data? choropleth or maybe bubble chart on top of a map
- crosstab? comparing two grouping variables with many level each? heatmap (perhaps with dendrograms in the margins)

````python
#=================================================-
#### Slide 20: Bivariate plots: scatter plot  ####

plt.figure(figsize=(6,4))         #<- set the figure size

keywords = dict(
    data=penguins,
    x = "bill_length_mm",          #<- set x variable
    y = "flipper_length_mm",        #<- set y variable
    linewidth = 0,
    )

# Create a scatterplot
sns.scatterplot(
    **keywords
    )                 #<- set line width


````

````python
sns.lmplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    hue="species", col="sex",
)
````

````python
#=================================================-
#### Slide 21: Bivariate plots: line plot  ####

plt.figure(figsize=(5,4))      #<- set the figure size

sns.lmplot(data = penguins,          #<- set the data
        x = "bill_length_mm",        #<- set x variable
        y = "flipper_length_mm",     #<- set y variable
        fit_reg = True,
        hue="species")


````

````python
sns.pairplot(
    data=penguins,
    hue="species",
    kind="reg",
    # palette="twilight",
    )
````

````python
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="magma",
    s=100,
    # kind="kde"
    
)
````

````python
sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="magma",
    s=100,
    kind="kde"
    
)
````

````python
g = sns.JointGrid(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="magma",
)
g.plot_joint(sns.scatterplot, s=100)
g.plot_marginals(sns.histplot, kde=True)
````

````python
g= sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    s=100,
)
g.plot_joint(sns.kdeplot, zorder=0, levels=6)
g.plot_marginals(sns.rugplot, height=.1)
````

````python
#=================================================-
#### Slide 22: Bivariate plots: residual plot  ####

plt.figure(figsize=(6,4))          #<- set the figure size
sns.residplot(data = penguins, 
        x = 'bill_length_mm',               #<- set x variable
        y = 'flipper_length_mm',            #<- set y variable
        lowess = True, 
        # color="b"
        )


````

````python
#=================================================-
#### Slide 23: Multivariate plots: Heatmaps  ####

plt.figure(figsize=(6,4))            #<- set the figure size
sns.heatmap(penguins.select_dtypes("number").corr(),         #<- set the data to find out the correlation
        vmin = -1, 
        vmax = 1, 
        center = 0,
        annot = True,
        square=True)


````

````python
#=================================================-
#### Slide 23: Multivariate plots: Heatmaps  ####
import numpy as np
plt.figure(figsize=(6,4))            #<- set the figure size+
corr = penguins.select_dtypes("number").corr()
g=sns.heatmap(
    corr.iloc[1:, :-1],
    center=0,
    annot=True,
    square=True,
    mask=np.triu(corr)[1:,:-1],
    annot_kws={"fontsize":14},
    # cmap="viridis",
    vmin = -1, 
    vmax = 1, 
    
).tick_params(labelsize=12)
plt.xticks(rotation=20, ha="right", rotation_mode="anchor")
# plt.yticks(rotation=45, ha="right", rotation_mode="anchor")

````

````python
#=================================================-
#### Slide 25: Format plots in seaborn (cont'd)  ####

plt.figure(figsize=(5,4))            #<- set the figure size
g = sns.lmplot(data = penguins,
        x = "bill_length_mm",                   #<- set x variable
        y = "flipper_length_mm",                #<- set x variable
        height = 4,                             #<- set the height
        fit_reg = True)
g.set(title = 'Bill Length versus Flipper Length',           #<- set the title
        xlabel = 'Bill Length (mm)',                         #<- set label for x variable
        ylabel = 'Flipper Length (mm)')                      #<- set label for y variable


````
